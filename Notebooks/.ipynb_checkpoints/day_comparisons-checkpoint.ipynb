```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import netCDF4 as nc
```

import functions defined on diferent notebook

```python
from ipynb.fs.full.Functions import *
```

extract the variables from netCDF files

```python
path0 = "/Users/jenifervivar/Desktop/NERTO_2022/LIDAR_JUN17-30_2021/FXD20210717/"
variables = VAR_EXTRACTOR(path0)

variables.extract()
cnr_17 = variables.cnr_day
time_17 = pd.to_datetime(variables.time_day, unit = 's', utc = True)

#create Data frame

july17_cnr = dataframe_set(cnr_17, time_17)


jul17_cnrstd = pbl_heigth(july17_cnr, 17)

#Getting liDAR derived values for PBL

pbl_17 = variables.atm_structures
df_structures_17 = dataframe_set(pbl_17, time_17)
lidar_pbl17 = pbl_lidar(df_structures_17)

lidar_pbl17 = lidar_pbl17[lidar_pbl17.index.date ==pd.Timestamp("2021-07-17").date()]
```

```python
plt.plot(jul17_cnrstd["time"].dt.hour, jul17_cnrstd['PBL_H_CNR'], label = "CNR derived Values")
plt.plot(lidar_pbl17.index.hour,lidar_pbl17, label = "LiDAR Values")
plt.title("PBL Calculations on July 17, 2021")
plt.ylabel("Heigth (m)")
plt.xlabel("Time (UTC)")
plt.legend()
#DateFormatter('% Y-% m-% d % H:% M:% S')
plt.show()
```
